```python
import pandas as pd

unemp_county =  pd.read_csv("datasets/output.csv")
df = pd.read_csv("datasets/min_wage.csv",encoding="latin")
minwage = pd.DataFrame()

for name,group in df.groupby("State"):
    if minwage.empty:
        minwage = group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name})
    else:
        minwage = minwage.join(group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name}))

# test_pd = df.groupby("State")
# alabama_df = test_pd.get_group("Alabama")
# #alabama_df.drop("Footnote",axis=1,inplace=True)
# test_df = alabama_df.copy()
# test_df.drop("Footnote",axis=1,inplace=True)
# test_df.head()
minwage.head()
```

```python
#Impute missing data
import numpy as np

minwage = minwage.replace(0, np.NaN).dropna(axis=1)
minwage.head()
unemp_county.head()
```

```python
def get_min_wage(year, state):
    try:
        return minwage.loc[year][state]
    except:
        return np.NaN
```

```python
get_min_wage(2012,"Colorado")
```

```python
unemp_county['min_wage'] = list(map(get_min_wage,unemp_county["Year"],unemp_county["State"]))
```

```python
#Presidential Results
pres16 = pd.read_csv("datasets/pres16results.csv")
pres16.head(15)
```

```python
top_candidates = pres16.head(10)['cand'].values
print(top_candidates)
```

```python
county_2015 = unemp_county[ (unemp_county['Year']==2015) & (unemp_county["Month"]=="February")]
county_2015.head()
```

```python
state_abbv = pd.read_csv("datasets/state_abbv.csv", index_col=0)
state_abbv.head()
```

```python
state_abbv_dict = state_abbv.to_dict()['Postal Code']
```

```python
county_2015['State'] = county_2015['State'].map(state_abbv_dict)
county_2015.tail()
```

```python
pres16.rename(columns={"county": "County", "st": "State"}, inplace=True)
pres16.head()
```

```python
for df in [county_2015, pres16]:
    df.set_index(["County", "State"], inplace=True)

pres16 = pres16[pres16['cand']=="Donald Trump"]
pres16 = pres16[['pct']]
pres16.dropna(inplace=True)
pres16.head(2)

```

```python
county_2015.head(2)
```

```python
all_together = county_2015.merge(pres16, on=["County", "State"])
all_together.dropna(inplace=True)
all_together.drop("Year", axis=1, inplace=True)
all_together.head()
```
