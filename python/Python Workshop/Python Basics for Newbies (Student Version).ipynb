### 1.5. List Slicing

List Slicing is the method of splitting a subset of a list.

##### Syntax: Listname[start:stop:step]

You can specifiy where to start and where to end the slicing and specify the step using the slicing operator <span style="color:blue"><b> : </b></span>

```python
# List slicing

x = [80, 81, 82, 83, 84, 85, 86, 87, 88]

print(x[1:3])     # [81, 82]                                                 

print(x[:])       # [80, 81, 82, 83, 84, 85, 86, 87, 88]   # Omitting the 'start' index starts the slice from the index 0.
print(x[::])      # [80, 81, 82, 83, 84, 85, 86, 87, 88]

print(x[0:3])     # [80, 81, 82]
print(x[:3])      # [80, 81, 82]

print(x[2:])      # [82, 83, 84, 85, 86, 87, 88]           # Omitting the 'stop' index extends the slice to the end of the list.
print(x[-7:-2])   # [82, 83, 84, 85, 86]
print(x[2:-5])    # [82, 83]
```

### 1.6. What is a Tuple?
A tuple is an ordered collection of elements. It is identical to a list in all aspects, except:
- Tuples are defined by enclosing the elements in parentheses.
- Tuples are immutable (means you can't add, delete or change items after the tuple is defined).

```python
x = ('Bing', 'Bang', 'Boom', 'Bingy', 'Bangy', 'Boo')

print(x)                              # ('Bing', 'Bang', 'Boom', 'Bingy', 'Bangy', 'Boo')

print(x[1])                           # 'Bang'

print(x[1::2])                        # ('Bang', 'Bingy', 'Boo')

x[2:4] = ["Merry", "Christmas"]       # TypeError: 'tuple' object does not support item assignment
print(x)                              # Remember: 'tuple' objects are immutable.
```

### 1.8. What is a Dictionary?
A dictionary is a set of unordered key-value pairs enclosed within curly bracket and separated by comma.

- Dictionaries are unordered (cannot be accessed by referring to an index). 
- Dictionaries are mutable (means you can add, delete or change items in a dictionary).
- Dictionaries can be accessed by keys. 

###### Keys must be unique. Duplicates are automatically removed by selecting the last value. 

```python
d = {'name': 'Yvette', 'age': 45, 'name': 'Rex'}           
print(d)                                                 # {'name': 'Rex, 'age': 45}
```

###### Keys must be immutable type  such as numbers, strings,or tuples. 

```python
d = {'name': 'Yvette', 'age': 45, [2, 2]: 25}
print(d)                                                 # TypeError: unhashable type: 'list.
                                                         # [2,2] is a list, which is mutable and cannot be used as a key. 
```

```python
d = {'name': 'Yvette', 'age': 45, (2, 2): 25}
print(d)                                                 # {'name': 'Yvette', 'age': 45, (2, 2): 25}
                                                         # (2, 2) is a tuple, which immutable and can be used as a key. 
```

###### Value can be of any type. 

```python
d = { 'a' : [1,2,3] , 'b' : {4,5,6} }
print(d)                                                 # {'a': [1, 2, 3], 'b': {4, 5, 6}}
```

### 3.1. What is a Function?
A function is a group of related statements that perform a specific task. 



##### Create a Function
Use the <b>def</b> keyword to define a Python function. 

```python
def greet(name):                            # This creates a function called 'greet' with an argument called 'name'.
    print("Hello, " + name + "!")
    print("How are you doing today?")
    
print(type(greet))
```

After the <b>def</b> has run, call the function by adding parentheses after the function's name.

```python
greet("Rex")                               # The value of the argument (i.e. 'Rex') is copied inside the function.
# Hello, Rex!
# How are you doing today?
```

##### Pass two arguments
The values of the arguments are copied to their corresponding parameters inside the function.

```python
def employee(name, job):
    print(name, 'is a', job)

employee('Rex', 'Software Developer')
# Rex is a Software Developer
```

##### Return Statement
The return statement is used to return a value from a function.

```python
def calculate(x, y):
    return x + y, x*y
    
result = calculate(5, 10)
print(result)                               # (15, 50)
```

### 3.2. Types of Arguments:
- Positional Arguments
- Keyword Arguments
- Default Arguments
- Varible Length Arguments

##### Positional Arguments
Values are copied to their corresponding parameters in order.

```python
def employee(name, job):
    print(name, 'is a', job)

employee('Rex', 'Software Developer')                   # Rex is a Software Developer       
```

```python
def employee(name, job):
    print(name, 'is a', job)

employee('Software Developer', 'Rex')                   # Software Developer is a Rex (Wrong Position!)       
```

##### Keyword Arguments:
Arguments are matched by name, not by position. 

```python
def employee(name, job):
    print(name, 'is a', job)

employee(name = 'Rex', job = 'Software Developer')     # Rex is a Software Developer
```

```python
def employee(name, job):
    print(name, 'is a', job)

employee(job = 'Software Developer', name = 'Rex')     # Rex is a Software Developer
```

```python
# Practice your Python skills! Do not copy!
```

##### Default Argument
Specify default values for arguments when defining a function. 

```python
def employee(name, job = 'Software Developer'):
    print(name, 'is a', job)

employee('Rex')                                       # Rex is a Software Developer
```

```python
def employee(name, job = 'Software Developer'):
    print(name, 'is a', job)

employee('Rex', 'Software Engineer')                 # Rex is a Software Engineer
```

##### Variable Length Arguments:
To create functions that take unlimited number of arguments using *args. 
The * collects all the unmatched positional arguments into a tuple. 

```python
def employee_age(*args):
    print(args)

employee_age(25, 27, 35, 28, 40)                     # (25, 27, 35, 28, 40)


def employee_name(*args):
    print(args)

employee_name("Bing", "Bang", "Boo")                 # ("Bing", "Bang", "Boo")
```

### 3.3. Variable Scope
Variable Scope is a region in a program where a variable is declared and used. Python has two main types of variable scopes:
- Local Scope: A variable declared within a function.
- Global Scope: A variable declared outside all functions. 



```python
def func():
    x = 50
    print("value inside function: ", x)

x = 100
func()
print("value inside function: ", x)           
```

```python
# Practice your Python skills! Do not copy!
```

##### Local variables are removed from memory when the function call exits.

```python
del x

def func():
    x = 50        # local scope x
    print(x)

func()            # 50

print(x)          # NameError: name 'x' is not defined
```

```python
# Practice your Python skills! Do not copy!
```

##### Global variables are accessible inside or outside of functions. 

```python
x = 88            # global scope x

def func():
    print(x)      # x is 88 inside def

func()
print(x)          # x is 88 outside def
```

### 3.4. Lamda Function
Lambda functions are blocks of code that can be assigned to variables, passed as an argument, or returned from a function call, in languages that support high-order functions. 

Programming languages that support lambda functions include Smalltalk, Lisp, Ruby, Scala, Python, and, more recently, Java and JavaScript.

Also known as anonymous function, i.e. requires no name.

Frequently used in conjunction with higher-order functions, which take functions as arguments and return functions.



##### Basic Examples
A lambda function is a single-line function declared with no name.
- It can have <b>any number of arguments</b>, but can only have <b>one expression</b>. 
- It behaves similarly to the function declared using the <b>def</b> keyword.

```python
def doubler(num):
    return num*2

print(doubler(4))                                               #8
```

```python
doubler = lambda num: num*2

print(doubler(4))                                               #8
```

```python
multiplier = lambda num1, num2: num1*num2

print(multiplier(5,2))                                          #10
```

```python
addition = lambda num1, num2, num3: num1 + num2 + num3

print(addition(4, 5, 6))                                        #15
```

```python
# Practice your Python skills! Do not copy!
calc = lambda num1,num2,num3: num1*num2*num3
print(calc(1,2,3))
```

##### Common Usage

When to use lambda functions?
- Within built-in functions, such as map() and filter()
- Customize the internal algorithm of key functions, such as sorted(), nlargest(), and nsmallest()

###### Map() Function:
- Returns a list of the results after applying the given function to each item of a given iterable (list, tuple etc.)
- Syntax: <b>map(function, iterable)</b>

```python
animals = ['cat', 'dog', 'cow']

upperCase = list(map(lambda x: x.upper(), animals))

print(upperCase)
```

```python
# Practice your Python skills! Do not copy!
plants = ['rose','daisies','lilies']

plants = list(map(lambda plant:plant.upper(),plants))
print(plants)
```

###### Filter() Function:
- Creates a list where the items are filtered through a function to test if the item is accepted or not.
- Syntax: <b>filter(function, sequence)</b>

```python
animals = ['cat', 'dog', 'cow']

hasLetterO = list(filter(lambda x: 'o' in x, animals))

print(hasLetterO)
```

```python
# Practice your Python skills! Do not copy!
plants = list(map(lambda plant:plant.lower(),plants))
hasLetterO = list(filter(lambda x:'o' in x,plants))
print(hasLetterO)
```

### 5.3. Files Handling
Python has several functions for creating, reading, updating and deleting files. 

##### Open() function
The key function for working with files in Python is the open() function.

The open() function takes two parameters: filename, and mode. 

There are four different methods (modes) for opening a file:
- "r" - Read - Default value. Opens a file for reading, error if the file does not exist
- "a" - Append - Opens a file for appending, creates the file if it does not exist
- "w" - Write - Opens a file for writing, creates the file if it does not exist
- "x" - Create - Creates the specified file, returns an error if the file exists

In addition you can specify if the file should be handled as binary or text mode
- "t" - Text - Default value. Text mode
- "b" - Binary - Binary mode (e.g. images)
